```python
import matplotlib.pyplot as plt

# Models and corresponding metrics
models = [
    "VGG16", "VGG19", "MobileNetV3Large", "ResNet50V2", "InceptionResNetV2",
    "DenseNet121", "DenseNet169", "DenseNet201", "InceptionV3", "EfficientNetB0", "Xception"
]

accuracy = [0.9208, 0.9224, 0.8053, 0.9492, 0.9316, 0.9586, 0.9379, 0.9676, 0.9420, 0.4635, 0.9610]
f1_score = [0.9210, 0.9224, 0.8038, 0.9489, 0.9317, 0.9584, 0.9377, 0.9676, 0.9420, 0.2964, 0.9610]
auroc = [0.9801, 0.9738, 0.8898, 0.9891, 0.9802, 0.9898, 0.9913, 0.9926, 0.9888, 0.6922, 0.9922]
mcc = [0.8420, 0.8441, 0.6157, 0.8977, 0.8616, 0.9160, 0.8793, 0.9346, 0.8856, 0.0000, 0.9211]
gmean = [0.9234, 0.9223, 0.8017, 0.9435, 0.9312, 0.9540, 0.9359, 0.9654, 0.9435, 0.4943, 0.9604]

# Plot
plt.figure(figsize=(14, 6))
plt.plot(models, accuracy, marker='o', label='Accuracy')
plt.plot(models, f1_score, marker='s', label='F1-Score')
plt.plot(models, auroc, marker='^', label='AUROC')
plt.plot(models, mcc, marker='D', label='MCC')
plt.plot(models, gmean, marker='x', label='G-Mean')

# Aesthetics
plt.xticks(rotation=90)
plt.title('Performance Comparison of Models')
plt.xlabel('Model')
plt.ylabel('Metric Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
